```python
y = 2 * x + 4
```

```python
noise = np.random.normal(0, 1, size=(len(x)))
```

```python
y += noise
```

```python
import sklearn.linear_model as linear_model
```

```python
import sklearn.model_selection as selection
```

```python
clf1 = linear_model.LinearRegression()
```

```python
X_train, X_test, y_train, y_test = selection.train_test_split(x, y)
```

```python
clf1.fit(X_train.reshape(-1, 1), y_train)
```

```python
import sklearn.metrics as metrics
```

```python
predictions = clf1.predict(X_test.reshape(-1, 1))
y1 = clf1.predict(x.reshape(-1, 1))
```

```python
metrics.mean_squared_error(predictions, y_test)
```
